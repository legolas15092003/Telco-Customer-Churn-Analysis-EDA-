* here we will create a function ie:-'conv' in which we will convert the value 0 to 'not present' and 1 to 'present'

```python
# here we will create a function ie:-'conv' in which we will convert the value 0 to 'not present' and 1 to 'present'

def conv(values):
  if values == 1:
    return 'present'
  elif values == 0:
    return 'not present'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

* here we have a comparision of SeniorCitizens :-
(It tells us what percentage of senior and non-senior customers churned or stayed.
Instead of raw numbers, it shows how likely each group is to leave.)

* (Here Comparative a Great Percentage Of People in SeniorCitizen Category have churned)

```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar
ax = ct_percent.plot(
    kind='bar',
    stacked=True,
    figsize=(8,6)
)

# Add percentage labels
for i in range(len(ct_percent)):
    for j in range(len(ct_percent.columns)):
        value = ct_percent.iloc[i, j]
        if value > 0:
            ax.text(
                i,
                ct_percent.iloc[i, :j+1].sum() - value/2,
                f"{value:.1f}%",
                ha='center',
                va='center',
                color='white',
                fontweight='bold'
            )

# Formatting
plt.title("Churn Percentage by Senior Citizen")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

* It shows how long customers stay (tenure) and how churn is distributed across those time periods — i.e., at which months customers are more likely to leave.

* (People who have used our services  for a long time have stayed and peoples  who have used our services for 1-2 monnths have churned   )

* The analysis indicates that customers who do not subscribe to value-added services such as Online Security, Tech Support, Device Protection, and Backup exhibit significantly higher churn compared to those who do.
Fiber-optic internet users show a noticeably higher churn rate than DSL users, suggesting possible price or service-quality sensitivity in that segment.
Customers with multiple phone lines and bundled services tend to remain more loyal, reflecting higher engagement and switching costs.
Streaming services (TV and Movies) are associated with better retention, as customers using these services churn less frequently.
Overall, service bundling and premium add-ons play a critical role in reducing customer churn and increasing long-term retention.

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set figure size
plt.figure(figsize=(18, 14))

for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(x=col, data=df, hue='Churn')
    plt.title(col)
    plt.xticks(rotation=15)
    plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
